# Bag of words 

```python
# We study the bag-of-words of a document from wikipedia
```

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.corpus import words
```

```python
import wikipedia
```

```python
page= wikipedia.page("Quantum mechanics")
```

```python
text=page.content
text[0:500]
```

# Text preprocessing

```python
word=word_tokenize(text.lower())
```

```python
Eng_words=words.words()
```

```python
# Remove non-English words
word=[w for w in word if w in Eng_words]
```

```python
# Remove stopwords
word=[w for w in word if w not in stopwords.words('english')]
```

```python
# Remove non-alphabet 
word=[w for w in word if w.isalpha()]
```

```python
import re
```

```python
# Remove single characters
text=re.sub(r'\b\w\b',''," ".join(word))
```

```python
word=word_tokenize(text.lower())
```

# Count the number of words in the text

```python
from collections import Counter
```

```python
Count_word=Counter(word)
```

```python
# Keep the 20 most common words
common_word=Count_word.most_common(20)
```

```python
label=[]
value=[]
for i in range(20):
    label.append(common_word[i][0])
    value.append(common_word[i][1])
```

```python
# Plot the most common words using bar chart
indexes=np.arange(len(label))
plt.figure(figsize=(20,20))
plt.bar(indexes,value)
plt.xticks(indexes,label)
plt.xticks(fontsize=12,rotation=45)
```
